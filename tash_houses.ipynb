# Data cleaning
1. Dropping unnecessary columns

```python
# changing types of numeric columns to one float type
# 1   rooms int64 -> float64 
# 2   size object -> float64 
# 3   level int64 -> float64 
# 4   max_levels int64 -> float64 
# 5   price object -> float64 

dataframe['rooms'] = dataframe['rooms'].astype(float)
dataframe['level'] = dataframe['level'].astype(float)
dataframe['max_levels'] = dataframe['max_levels'].astype(float)
```

```python
dataframe['size'] = pd.to_numeric(dataframe['size'], errors='coerce')
```

```python
dataframe['price'] = pd.to_numeric(dataframe['price'], errors='coerce')
```

```python
dataframe.info()
```

```python
# removing null values of all columns
dataframe.describe()
```

```python
dataframe.isnull().sum()
```

```python
dataframe2 = dataframe.dropna(axis=0,how='any')
```

```python
dataframe2.isnull().sum()
```

```python
# removing unusual values
dataframe2.describe()
```

```python
min_size_threshold = 25  # Example minimum price threshold
max_size_threshold = 200  # Example maximum price threshold

dataframe2 = dataframe2[(dataframe2['size'] > min_size_threshold) & (dataframe2['size'] < max_size_threshold)]
```

```python
dataframe2.describe()
```

```python
# We should remove location and distract columns for correlation
dataframe3 = dataframe2.drop(['location', 'district'], axis=1)
```

# Correlation and visualisations

```python
# correlation with numbers
plt.figure(figsize=(14,7))
sns.heatmap(dataframe3.corr(),annot=True)
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
sns.heatmap(data=dataframe3.corr(), linewidths=0.5, center=0, annot=True, ax=ax)
```

```python
correlation_matrix = dataframe3.select_dtypes(include='number').corr()
high_corr = correlation_matrix[(correlation_matrix > 0.01) & (correlation_matrix != 1)]
high_corr = high_corr.stack().reset_index()
high_corr.columns = ['level_1', 'level_2', 'Correlation']
high_corr = high_corr.dropna()
high_corr
```
